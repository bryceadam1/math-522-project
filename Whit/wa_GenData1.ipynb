```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.random.uniform(0,2*np.pi,size=1000)
```

```python
y = np.sin(x) + np.random.normal(scale=0.25, size=1000)
```

```python
dom = np.linspace(0,2*np.pi,1000)
plt.scatter(x,y)
plt.plot(dom,np.sin(dom),'r')
plt.show()
```
